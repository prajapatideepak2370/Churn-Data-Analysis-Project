```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Customer Churn.csv")
df.head()
```

```python
df.info()
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum()
df.isna().sum().sum()
```

```python
df.duplicated().sum()
```

```python
def conv(Value):
    if Value == 1:
        return "yes"
    else:
        return "no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
df.head()
```

```python
plt.figure(figsize= (4,5))
ax = sns.countplot(x="Churn", data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of customers in churn")
plt.show()
```

```python
plt.figure(figsize = (3,4))
pi = df.groupby("Churn").agg({'Churn' : "count"})
plt.pie(pi['Churn'], labels = pi.index,autopct= "%1.2f%%")
plt.title("Percentage of customers in churned")
plt.show()
```

```python
plt.figure(figsize = (5,6))
gen = sns.countplot(x = "gender", data=df, hue = "Churn")
gen.bar_label(gen.containers[0])
gen.bar_label(gen.containers[1])
plt.title("Count of Gender in Churn")
plt.show()
```

```python
plt.figure(figsize=(5,6))
ser = sns.countplot(x="SeniorCitizen", hue="SeniorCitizen", data=df,  legend=True)
ser.bar_label(ser.containers[0])
ser.bar_label(ser.containers[1])
plt.title("Count of Senior Citizen in Churn")
plt.show()
```

```python
total_counts = df.groupby("SeniorCitizen")["Churn"].value_counts(normalize=True).unstack()

fig, ax = plt.subplots(figsize= (8,7))
total_counts.plot(kind = "bar", stacked = True, ax=ax)
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x,y = p.get_xy()
    ax.text(x + width/2, y + height/2, f'{height:.1f}%', ha="center", va="center")

plt.title("Churn by  Senior Citizen(Stacked Bar Chart)")
plt.xlabel("SeniorCitizen")
plt.ylabel("Percentage (%)")
plt.xticks(rotation = 0)
plt.legend(title = 'Churn', loc = 'upper right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10,8))
ten = sns.histplot(x = 'tenure',
                   data = df,
                   legend = True,
                   hue = "Churn",
                   bins =40)
ten.bar_label(ten.containers[0])
ten.bar_label(ten.containers[1])
plt.tight_layout()
plt.title("")
plt.show()
```

```python
plt.figure(figsize=(5,6))
ser = sns.countplot(x="Contract", data=df, hue = "Churn",  legend=True)
ser.bar_label(ser.containers[0])
ser.bar_label(ser.containers[1])
plt.title("Count of Customer  by Contract in Churn")
plt.show()
```

```python
df.columns.values
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# figure size depends on number of subplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15,12))
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.countplot(x=col, hue=df["Churn"], data=df, palette="Accent_r", legend=True, ax=axes[i])
    axes[i].set_title(f"Count of {col}")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")
    # add labels
    for container in axes[i].containers:
        axes[i].bar_label(container, fmt='%d')

plt.tight_layout()
plt.show()
```

Customer is likely to Churn when he is using electronic check as a payment method.

```python
plt.figure(figsize=(7,4))
ser = sns.countplot(x="PaymentMethod", data=df, hue = "Churn",  legend=True)
ser.bar_label(ser.containers[0])
ser.bar_label(ser.containers[1])
plt.xticks(rotation = 45)
plt.title("Churn Customer by Payment method")
plt.show()
```
